```python
import sqlite3

conn = sqlite3.connect("./books.db")
cur = conn.cursor()
```

```python
cur.execute("SELECT * FROM books")
db_list = list(cur.fetchall())

for i in db_list:
    print(i)
    print("\n========================================\n")
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
from konlpy.tag import Hannanum
```

```python
hannanum = Hannanum()
nouns = []
```

```python
for i in db_list:
    if i[1] == "4":
        nouns += hannanum.nouns(i[5])
```

```python
counter = Counter(nouns)
new_counter = {}

for k, v in counter.items():
    if v > 5 and len(k) > 1:
        new_counter[k] = v
```

```python
wordcloud = WordCloud(font_path="/Users/Song/fonts/tree.ttf", background_color="white", width=2000, height=2000)
img = wordcloud.generate_from_frequencies(new_counter)
plt.imshow(img)
```
