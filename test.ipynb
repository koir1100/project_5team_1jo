```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('./test.db')
cur = conn.cursor()
cur.execute('SELECT * FROM book_keyword')
```

```python
check_list = [i for i in cur if i[2] == 1]

check_list
```

```python
x_list = [i[1] for i in check_list]
y_list = [i[3] for i in check_list]
```

```python
dict_check = {}
```

```python
for i in y_list:
    for j in i.split(','):
        print(j)
        dict_check[j] = dict_check.get(j, 0) + 1
```

```python
dict_check
```

```python
sns.lineplot(x=dict_check.keys(), y=dict_check.values())
```

```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud()
```

```python
img = wordcloud.generate_from_frequencies(dict_check)
```
